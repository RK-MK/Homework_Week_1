## Bias Variance Tradeoff

In machine learning we try to build a model that fits our data. A model with a low complexity leans towards a low variance but this usually leads to a high bias and vice versa.
- an algorithm with high bias leans to miss the relevant relations between features and target ouputs (underfitting)
- an algorithm with high variance leans to model the random noise of the training data, rather than the intended outputs (overfitting)
- So our goal is to find a complexity for our model that leans to a low bias and a low variance to have a minimum total error
